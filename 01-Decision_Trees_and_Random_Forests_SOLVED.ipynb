# Decision Trees and Random Forests in Python

Using the **Kyphosis** dataset to demonstrate Decision Trees and Random Forests.

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Get the Data

```python
df = pd.read_csv('kyphosis.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## EDA

A pairplot to visualize relationships between features, colored by Kyphosis outcome.

```python
sns.pairplot(df, hue='Kyphosis', palette='Set1')
plt.suptitle('Pairplot of Kyphosis Dataset', y=1.02)
plt.show()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

X = df.drop('Kyphosis', axis=1)
y = df['Kyphosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
print(f"Training samples: {X_train.shape[0]}")
print(f"Testing samples:  {X_test.shape[0]}")
```

## Decision Trees

Train a single Decision Tree classifier.

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
```

## Prediction and Evaluation – Decision Tree

```python
from sklearn.metrics import classification_report, confusion_matrix

predictions = dtree.predict(X_test)

print("=== Classification Report ===")
print(classification_report(y_test, predictions))

print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, predictions))
```

## Random Forests

Now let's compare the Decision Tree to a Random Forest with 100 estimators.

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)

print("=== Confusion Matrix – Random Forest ===")
print(confusion_matrix(y_test, rfc_pred))

print()
print("=== Classification Report – Random Forest ===")
print(classification_report(y_test, rfc_pred))
```
